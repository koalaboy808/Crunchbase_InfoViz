```python
agg_df = _df.groupby(["Investor"]).sum().reset_index()
```

```python
agg_df = agg_df.drop('Date', 1)
```

```python
keep_countries = ['USA', 'GBR', 'CAN', 'CHN', 'DEU', 'ISR', 'FRA', 'CHE', 'JPN', 'RUS']
subset_df = agg_df[agg_df['Investor'].isin(keep_countries)]
```
