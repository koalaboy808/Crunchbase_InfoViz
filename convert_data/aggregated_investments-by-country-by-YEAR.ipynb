```python
df = pd.read_csv("../raw_data_crunchbase/crunchbase-investments.csv")
# df = df[df['funded_year']==2013]
# df['investor_permalink'] = df['investor_permalink'].apply(lambda x: split_backslash(x))
# df = df.loc[df['investor_permalink']=='list_unicornsorganization'] # change to person!!!!!!!

# from 20179 to 17617 for raised_amount_usd
df = df.loc[df["raised_amount_usd"].notnull()]
df['raised_amount_usd'] = df['raised_amount_usd'].apply(lambda x: "".join(x.split(",")).strip())
df.loc[df.raised_amount_usd == '-', 'raised_amount_usd'] = None

df['raised_amount_usd'] = df['raised_amount_usd'].map(lambda x: float(x) if x is not None else x)

# # all present for funding_round_type
# df = df.loc[df["funding_round_type"].notnull()]
# # from 17617 to 16553 for company_country_code
df = df.loc[df["investor_country_code"].notnull()]
df = df.loc[df["investor_city"].notnull()]
df = df.loc[df["company_city"].notnull()]
df = df.loc[df["company_country_code"].notnull()]
```

```python
subset_df = df[["company_name", "company_country_code", "company_city", "company_state_code",
                "investor_name","investor_country_code","investor_city", "investor_state_code",
                "funded_month","funded_year","raised_amount_usd"]]
```

```python
years_wanted = [2004,2005,2006,2007,2008,2009,2010,2011,2012,2013]
```

```python
# top25 = ["E-Commerce","Advertising","Software","Storage","Clean Technology","Communities","File Sharing","Technology","Online Shopping","Curated Web","Enterprise Software","Design","Games","Cloud Computing","Shopping","Entertainment","Transportation","Security","Data Security","Peer-to-Peer","Mobile Payments","Search","Fashion","Facebook Applications","Finance Technology","Consumer Electronics","Analytics","Privacy","Mobile Games","News","Hardware + Software","Retail","Mobile","Location Based Services","Biotechnology"]
# subset_df = subset_df[add_unicorns["company_market"].isin(top25)]
```

```python
subset_df = subset_df.loc[subset_df['funded_year'].isin(years_wanted)]
# subset_df = subset_df[subset_df['funded_year']>=2004 and subset_df['funded_year']>=2013]
```

```python
subset_df.head()
```

```python
# cum_df = subset_df.copy()
# cum_df = cum_df.sort(['funded_month','investor_country_code','company_country_code'], ascending=[1,1,1])
# cum_df['no_cumulative'] = cum_df.groupby(['funded_month','investor_country_code','company_country_code'])['raised_amount_usd'].apply(lambda x: x.cumsum())
# cum_df = cum_df.drop_duplicates(['funded_month','investor_country_code','company_country_code'], take_last=True)
# cum_df = cum_df[['funded_month','investor_country_code','company_country_code','funded_year','raised_amount_usd','no_cumulative']]
# cum_df = cum_df.sort(['funded_month'], ascending=[1])
# cum_df[:20]

# subset_df.groupby(['funded_month','investor_country_code','company_country_code']).sum().groupby(level=[0]).cumsum()
country_grouped = subset_df.groupby(['funded_year','investor_country_code','company_country_code']).sum().reset_index()[['funded_year','investor_country_code','company_country_code',"raised_amount_usd"]]
```

```python
country_grouped = country_grouped.sort(['funded_year','investor_country_code','company_country_code'], ascending=[1,1,1])
country_grouped['no_cumulative'] = country_grouped.groupby(['investor_country_code','company_country_code'])['raised_amount_usd'].apply(lambda x: x.cumsum())
country_grouped[:20]
```

```python
country_grouped.shape
```

```python
# country_grouped['_unique'] = country_grouped['funded_month'].apply(lambda x: )
country_grouped["_unique"] = country_grouped["funded_year"].map(str) + "_" + country_grouped["investor_country_code"]
country_grouped.head()
```

```python
topcountries = country_grouped[country_grouped['funded_year']==2013]
topcountries = topcountries.groupby(['investor_country_code']).sum().reset_index()[['investor_country_code',"no_cumulative"]]
print(list(topcountries.sort(['no_cumulative'], ascending=[0])['investor_country_code'][:15]))
# topcountries['no_cumulative'] = topcountries.groupby(['investor_country_code'])['no_cumulative'].apply(lambda x: x.cumsum())
# topcountries.sort(['no_cumulative','investor_country_code','company_country_code'], ascending=[1,1,1])
max(country_grouped["no_cumulative"])
```

```python
# country_wide = country_grouped.pivot(index='_unique', columns='company_country_code', values='raised_amount_usd')
country_wide = country_grouped.pivot(index='_unique', columns='company_country_code', values='no_cumulative')
country_wide = country_wide.fillna(0)
```
